### Base final (Graduados Total)

Después de analizar las variables de manera independiente, se procederá a graficar la base de datos final. Base de datos que incluye todas las variables, lo que permite realizar gráficos para observar las relaciones entre ellas. Este análisis visual facilita la identificación de patrones y tendencias, lo que a su vez permite obtener conclusiones más fundamentadas y comprensibles.

```python
Gradudados_final_stats = pd.read_excel("Graduados_total.xlsx")
```

Relación entre graduados y gocentes por municipio

```python
import matplotlib.pyplot as plt
graduados = Gradudados_final_stats.groupby('COD_MUN')[['GRADUADOS', 'DOCENTES']].mean()
tiempo = Gradudados_final_stats.groupby('AÑO')[['GRADUADOS', 'DOCENTES']].sum()
graduates_per_area = Gradudados_final_stats.groupby('AREA_CON')['GRADUADOS'].sum()

plt.figure(figsize=(10, 6))
plt.scatter(graduados['DOCENTES'], graduados['GRADUADOS'], alpha=0.7)
plt.title('Relación entre graduados y docentes por municipio (promedio)', fontsize=14)
plt.xlabel('Número Promedio de Docentes', fontsize=12)
plt.ylabel('Número Promedio de Graduados', fontsize=12)
plt.grid(True)
plt.show()
```

Existe una correlación positiva entre el número promedio de docentes y el número promedio de graduados. Esto indica que, en general, un mayor número de docentes en un municipio se asocia con un mayor número de graduados. Esto sugiere que la disponibilidad de docentes está correlacionada con el número de graduados.

Evolución de graduados, focentes y matriculados por año

```python
import matplotlib.pyplot as plt

evolucion = Gradudados_final_stats.groupby('AÑO')[['GRADUADOS', 'DOCENTES', 'MATRICULADOS']].sum()

plt.figure(figsize=(10, 6))
for column in evolucion.columns:
    plt.plot(evolucion.index, evolucion[column], marker='o', label=column)
plt.title('Evolución de graduados, focentes y matriculados por año', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

La cantidad de estudiantes matriculados se mantiene relativamente constante en todos los años, aunque tiene algunas variaciones variaciones. 
La cantidad de docentes muestra un incremento leve, lo que podría indicar esfuerzos por mejorar la disponibilidad de profesiones en el sector educativo a lo largo del tiempo.
Aunque los graduados muestran una tendencia estable, la cantidad sigue siendo significativamente menor en comparación con los matriculados.

Relación entre población y PIB

```python
plt.figure(figsize=(10, 6))
plt.scatter(Gradudados_final_stats['POBLACION'], Gradudados_final_stats['PIB'], alpha=0.6, edgecolor='k')
plt.title('Relación entre población y PIB', fontsize=14)
plt.xlabel('Población', fontsize=12)
plt.ylabel('PIB', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Los municipios con mayor población tienden a tener un PIB más alto. 

Mapa de correlación

```python
correlation_data = Gradudados_final_stats[['GRADUADOS', 'DOCENTES', 'MATRICULADOS', 'POBLACION', 'PIB']].corr()

plt.figure(figsize=(8, 6))
plt.matshow(correlation_data, cmap='coolwarm', fignum=1)
plt.title('Mapa de correlación', fontsize=14, pad=20)
plt.colorbar()
plt.xticks(range(len(correlation_data.columns)), correlation_data.columns, rotation=45, ha='left')
plt.yticks(range(len(correlation_data.columns)), correlation_data.columns)
plt.tight_layout()
plt.show()
```

Existe una correlación positiva fuerte entre los matriculados y los graduados, lo que es esperado, ya que un mayor número de estudiantes matriculados tiende a resultar en un mayor número de graduados. Por otro lado los docentes tienen una correlación significativa con los matriculados, lo que indica que una mayor disponibilidad de docentes está relacionada con un mayor número de estudiantes inscritos. Aunque el PIB tiene correlación con las variables educativas (graduados, docentes y matriculados), la relación no es tan fuerte como con la población.

Relación entre matriculados y graduados

Existe una correlación positiva fuerte entre los matriculados y los graduados, lo que es esperado, ya que un mayor número de estudiantes matriculados tiende a resultar en un mayor número de graduados.
